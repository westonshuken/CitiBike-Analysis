```python
sample_df['started_at'] = pd.to_datetime(sample_df['started_at'])
sample_df = sample_df.groupby([pd.Grouper(key='started_at', freq='D'), sample_df['rideable_type']+sample_df['member_casual'],]).agg({'ride_id': ['count']})
sample_df = sample_df.reset_index()
sample_df.columns = [i[0] for i in sample_df.columns]
sample_df.sort_values('started_at')
```

```python
# Daily Data Union for 2022 and 2023

path = ["/Users/westonshuken/Documents/_2025/CitiBike/Data/citibike_tripdata_2022", "/Users/westonshuken/Documents/_2025/CitiBike/Data/citibike_tripdata_2023"]

dfs = []

tick = 0

for i in path:
    print(i)
    for filename in os.listdir(i):
        tick += 1
        if filename.endswith('.csv'):
            file_path = os.path.join(i, filename)
            print("reading file: ", str(tick))
            df = pd.read_csv(file_path, dtype={'start_station_id': str, 'end_station_id': str})
            df['ended_at'] = pd.to_datetime(df['ended_at'])
            grouped_df = df.groupby([pd.Grouper(key='ended_at', freq='D'), df['rideable_type']+df['member_casual'],]).agg({'ride_id': ['count']})
            grouped_df = grouped_df.reset_index()
            grouped_df.columns = [i[0] for i in grouped_df.columns]
            dfs.append(grouped_df)
            #if tick >=1:
               #break
```

```python
concat_df = pd.concat(dfs, ignore_index=True)
```

```python
cols = {'ended_at': 'DATE', 'level_1': 'RIDE_TYPE', 'ride_id': 'RIDES'}
concat_df = concat_df.rename(columns=cols)
concat_df['DATE'] = pd.to_datetime(concat_df['DATE'])
```

```python
concat_df[concat_df['DATE'] <= '2022-01-01']
```

```python
concat_df.sort_values('DATE', inplace=True)
```

```python
concat_df = concat_df.reset_index(drop=True)
concat_df
```

```python
date_df = concat_df.groupby(['DATE','RIDE_TYPE']).sum('RIDES')
date_df = date_df.reset_index()
date_df.columns = [i for i in date_df.columns]
date_df
```

```python
climate_df = pd.read_csv("/Users/westonshuken/Documents/_2025/CitiBike/climatedata.csv")
```

```python
climate_df['DATE'] = pd.to_datetime(climate_df['DATE'])
climate_df.info()
```

```python
final_df = pd.merge(date_df, climate_df, on='DATE', how='left')
```
